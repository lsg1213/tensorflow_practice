```python
marathon_2015_2017 = pd.read_csv('./data/marathon_2015_2017.csv')

# training set
marathon_train = marathon_2015_2017[marathon_2015_2017['Year'] != 2017]
marathon_test = marathon_2015_2017[marathon_2015_2017['Year'] == 2017]

# record
record_train = pd.DataFrame(marathon_train, columns=['Age', 'M/F', 'Pace'])
record_test = pd.DataFrame(marathon_test, columns=['Age', 'M/F', 'Pace'])

def learning(age = 26, gender = 'M', rank = 1, step = 1, training = 100):
    model = tf.keras.models.Sequential()
    train = record_train[record_train['Age'] == age]
    train = train[train['M/F'] == gender].sort_values(['Pace'], ascending=[True])
    test = record_test[record_test['Age'] == age]
    test = test[test['M/F'] == gender].sort_values(['Pace'], ascending=[True])

    model.add(tf.keras.layers.Dense(1, input_shape=(3,)))
    
    sgd = tf.keras.optimizers.SGD(lr=step)
    
    model.compile(loss='mse',
                 optimizer=sgd,
                 metrics=['accuracy'])
    model.summary()
    history = model.fit(np.array(train[0:2]), np.array(train[2]), epochs=step)
    for i in range(training):
        print(i)
        if i % 10 == 0:
            print('{}\t {}'.format(i, history.history['loss'][i]))
    
    #testing
    pace = model.predict(np.array([age, gender]))
    print('predict: {}, real: {}'.format(pace[rank], test[rank] - pace[rank]))
```

```python
learning()
```
